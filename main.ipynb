```python
import matplotlib.pyplot as plt
import numpy as np

def positional_encoding(position, d_model):
    '''
    Compute positional encoding for a particular position
    '''
    # Initialize a positional encoding
    PE = np.zeros((1, d_model))
    
    # Compute the positional encoding
    for i in range(d_model):
        if i%2 == 0:
            PE[0][i] = np.sin(position/np.power(10000, i/d_model))
        else:
            PE[0][i] = np.cos(position/np.power(10000, (i-1)/d_model))
    
    return PE


def plot_positional_encoding(max_position, d_model):
    '''
    Compute and plot positional encoding for positions from 0 to max_position
    '''
    # Initialize a positional encodings matrix
    PEs = np.zeros((max_position, d_model))
    
    for pos in range(max_position):
        PEs[pos] = positional_encoding(pos, d_model)
    
    plt.figure(figsize=(12,8))
    plt.pcolormesh(PEs, cmap='viridis')
    plt.xlabel('Depth')
    plt.xlim((0, d_model))
    plt.ylim((max_position,0))
    plt.ylabel('Position')
    plt.colorbar()
    plt.show()


# Test
plot_positional_encoding(50, 512)
```

```python
def plot_positional_encoding_mutiply_transposed(max_position, d_model):
    '''
    Compute positional encoding for positions from 0 to max_position, 
    multiply by its transopse and plot the result
    '''
    # Initialize a positional encodings matrix
    PEs = np.zeros((max_position, d_model))
    
    for pos in range(max_position):
        PEs[pos] = positional_encoding(pos, d_model)
    
    # Multiply PEs by its transpose
    PEs_product = np.matmul(PEs, PEs.T) + np.eye(max_position)
    
    plt.figure(figsize=(12,8))
    plt.pcolormesh(PEs_product, cmap='viridis')
    plt.xlabel('Position')
    plt.xlim((0, max_position))
    plt.ylim((max_position,0))
    plt.ylabel('Position')
    plt.colorbar()
    plt.show()

# Test
plot_positional_encoding_mutiply_transposed(50, 512)
```
